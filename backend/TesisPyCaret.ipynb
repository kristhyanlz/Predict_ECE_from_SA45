```python
import pandas as pd
```

```python
url = "https://docs.google.com/spreadsheets/d/1vFWwNJ7-61O_QYaPo8d7bmw7K192az4R/export"
data = pd.read_excel(url, sheet_name=1)
```

```python
data.head()
```

```python
data = data.iloc[:, -10:]
data.head()
```

```python
category_map = {
    'Bajo': 0,
    'Promedio': 1,
    'Alto': 2,
    'En riesgo': 3
}
```

```python
data2 = data.copy()
for i in range(10):
    data2.iloc[:, i] = data2.iloc[:, i].map(category_map)
data2.head()
```

```python
data2.to_csv("respuestas_label.csv")
```
